```python
# 배열 List 1 과 list 2 를 각 set 자료형으로 설정합니다. 그리고 두 set 자료형의 교집합, 합집합, 차집합을 차례로
list1 = [1,2,3,4,5,6]
list2 = [4,5,6,7,8,9]

set1=set(list1)
set2=set(list2)
print(set1&set2)
print(set1|set2)
print(set1-set2)
```
