```python
hit = 0

while hit<10:
  hit+=1
  print('나무를 {}번 찍었습니다.'.format(hit))
  if hit ==10:
    print('나무가 쓰러졌습니다.')
```

```python
coffee = 10
while True:
  coffee = coffee - 1
  print('커피를 한잔팔고, {}잔 남았습니다.'.format(coffee))
  if coffee ==0:
    print('커피를 모두 팔았습니다.')
    break #반복문을 돌리다가 반복문을 강제로 나간다.
```

```python
a = 0

while a<10: # 여기로 이동합니다.
  a+= 1
  if a%2 == 0:
    continue # 반복문을 돌리다가 그 순간 
  print(a)
```
