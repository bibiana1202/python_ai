```python
def intro(name,age,sex = True):
  print("나의 이름은 {} 입니다.".format(name))
  print('나의 나이는 {} 입니다.'.format(age))
  if sex == True:
    print('나는 남자입니다.')
  else:
    print('나는 여자입니다.')
```

```python
intro('곽경일',30)
```

```python
intro('박혜정',30,False)
```

```python
def test():
  a = 5 # test 함수에 있는 a 에 5

a=0 # a 에 0
test()
a
```

```python
def triangle(height,base):
  area = (height ** 2 + base **2) ** 1/2
  return(area)
```

```python
print("직각삼각형의 빗변을 구하는 프로그램 입니다.")
h = int(input("직각삼각형의 밑변의 길이를 입력해 주세요:"))
b = int(input("직각삼각형의 높이의 길이를 입력해 주세요:"))
result = triangle(h,b)
print("입력하신 높이와 밑변의 길이의 직각 삼각형의 빗변의 길이는",result,"입니다.")
```
