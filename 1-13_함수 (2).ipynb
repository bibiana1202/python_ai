```python
# 5! = 5*4*3*2*1 =120

def fac(n):
  a=1
  for i in range(1,n+1):
    a*= i
  return a
```

```python
fac(5)
```

```python
# 재귀 함수 : 함수 내에서 자기 자신을 불러오는 함수

def fac(n):
  if n == 1:
    return 1
  else:
    return n * fac(n-1)
```

```python
fac(5)
```

```python
# 피보나치수열 : 앞에있는거 2개를 더하는 수열
# 1 1 2 3 5 8 13 ...

dic = {1:1,2:1}

count = 0

def pivo(n):
  global count
  count += 1
  if n in dic:
    return dic[n]
  else:
    dic[n] = pivo(n-1) + pivo(n-2)
    return dic[n]


  # 메모라이징 기법
```

```python
pivo(35)
```

```python
def fibo(n):
  if n ==1 or n ==2:
    return 1
  else:
    return fibo(n-1) + fibo(n-2)    
```
