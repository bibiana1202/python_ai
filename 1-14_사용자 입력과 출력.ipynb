```python
f = open('test.txt','w')

for i in range(1,11):
  f.write('{}번째 줄입니다.\n'.format(i))

f.close() # 저장하고 종료해주는 기능
```

```python
f = open('test.txt','a') # attach 이어 붙이다.

for i in range(1,11):
  f.write('{}번째 줄입니다.\n'.format(i))

f.close() # 저장하고 종료해주는 기능

```

```python
f = open('test.txt','r')

data = f.read()

f.close() # 저장하고 종료해주는 기능

data
```

```python
f = open('test.txt','r')

data = f.readlines()

f.close() # 저장하고 종료해주는 기능

data
```

```python
# 변수 a 에 5,b 에 3을 할당한 후 result.txt 메모장을 생성하여, 메모장 안에 a + b 결과값을 넣은후 저장해보시오.

a=5
b=3

f= open('result.txt','w')
f.write("{}".format(a+b))

f= open('result.txt','a')
f.write("{}".format(a+b))

f= open('result.txt','r')
data = f.read()
f.close()

data
```
