```python
class Bank: #클래스 : 함수들을 저장해놓은 주머니
  def __init__(self,a): # 권한을 줄때 자동으로 한번 실행되는 함수
    self.money = a
  def deposit(self,a):
    self.money += a
```

```python
person1 = Bank(50000) # 이시점!
person1.money
```

```python
person1.deposit(30000)
person1.money
```

```python
person2 = Bank()
person2.deposit(30000)
```

```python
person2.money
```

```python
class Bank2(Bank): #Bank2 는 Bank 에서 제공하는 서비스를 모두 상속 받는다.
  pass
```

```python
person3 = Bank2(20000)
```

```python
person3.money
```

```python
person3.deposit(60000)
person3.money
```
