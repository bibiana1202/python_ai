```python
import pandas as pd
import seaborn as sns
```

```python
anscombe = sns.load_dataset("anscombe")
```

```python
anscombe
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure() # 그림을 그리기 위한 기반을 마련 / 도화지를 쫙 편다!

ax1 = fig.add_subplot(2,2,1) # 자리 할당
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
data1 = anscombe[anscombe['dataset']=='I']
data2 = anscombe[anscombe['dataset']=='II']
data3 = anscombe[anscombe['dataset']=='III']
data4 = anscombe[anscombe['dataset']=='IV']
```

```python
data1
```

```python
ax1.scatter(data1['x'],data1['y'])
ax2.scatter(data2['x'],data2['y'])
ax3.scatter(data3['x'],data3['y'])
ax4.scatter(data4['x'],data4['y'])

fig
```

```python
print(data1.mean())
print(data2.mean())
print(data3.mean())
print(data4.mean())
```

```python
print(data1.std())
print(data2.std())
print(data3.std())
print(data4.std())
```

```python
ax1.set_title('data1')
ax2.set_title('data2')
ax3.set_title('data3')
ax4.set_title('data4')

fig
```

```python
fig.suptitle('Anscombe Data')
fig
```
