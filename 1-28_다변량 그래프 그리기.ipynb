```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips= sns.load_dataset("tips")
tips
```

```python
def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
```

```python
tips['sex_color']= tips['sex'].apply(recode_sex)
tips
```

```python
fig = plt.figure()
ax=fig.add_subplot(1,1,1)

ax.scatter(x=tips['total_bill'],
           y=tips['tip'],
           s=tips['size']*10,
           c=tips['sex_color'],
           alpha =0.5)
```
