```python
import pandas as pd

weather = pd.read_csv("weather.csv")

weather
```

```python
weather2 = pd.melt(weather,id_vars=['id','year','month','element'],var_name='day',value_name='temp')
weather2
```

```python
weather3 = weather2.pivot_table(
    index=['id','year','month','day'],
    columns = 'element',
    values  = 'temp',
    dropna = False
)
weather3
```

```python
weather3.reset_index()
```
