```python
import re

tel = '0123456789'

re.findall('\d\d\d\d\d\d\d\d\d\d',tel)
```

```python
# <정규 표현식>
# \d 숫자
# \s 공백
# ?  있을수도 있고 없을수도 있다.
tel='012 345 6789'
re.findall('\d{3}\s?\d{3}\s?\d{4}',tel)
```

```python
# () capture 딱 only 이부분만 찾고 싶다.

tel='(012) 345 6789'
re.findall('\(\d{3}\)\s?\d{3}\s?\d{4}',tel)
```

```python
tel='+1 (012) 345 6789'
re.findall('\+\d\s\(\d{3}\)\s?\d{3}\s?\d{4}',tel)
```
