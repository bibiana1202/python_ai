```python
import requests # 요청하다.
from bs4 import BeautifulSoup # 파이썬으로 HTML 다루는 라이브러리
import pandas as pd # 데이터 분석 툴
```

```python
url =requests.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EC%97%B0%EA%B8%88%EB%B3%B5%EA%B6%8C")
url
```

```python
# 200 정상
# url.text # HTML 처럼 생긴 문자열
html = BeautifulSoup(url.text) # 자동으로 HTML 변환
```

```python
box=[]
for i in html.select("div.winning_number")[0].select('span.ball'):
  box.append(int(i.text))


box
```
