```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = requests.get('https://finance.naver.com/sise/sise_market_sum.naver')
url
```

```python
html= BeautifulSoup(url.text)
table = html.select('table.type_2')[0]
table
```

```python
table=pd.read_html(str(table))[0]
table
```

```python
del table['N']
del table['토론실']
table
```

```python
table= table[table['종목명'].notnull()]
table
```

```python
table.to_excel('stock.xlsx')
```
