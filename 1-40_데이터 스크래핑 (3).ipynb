```python
# 네이버 뉴스 기사 제목과 내용을 가져와서 단어 빈도를 분석

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = requests.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EC%BD%94%EB%A1%9C%EB%82%98")
url
```

```python
html =BeautifulSoup(url.text)
```

```python
total=''

for i in html.select('ul.list_news')[0].select('li.bx'):
  title= i.select('a.news_tit')[0].text
  content = i.select('div.news_dsc')[0].text
  total += (title +' ' +content)

total 
```

```python
import re

words = re.findall('[가-힣]+',total)
words
```

```python
dic = {}

for i in words:
  if i not in dic:
    dic[i] = 1
  else:
    dic[i]+=1
  
dic
```

```python
df = pd.DataFrame({
    '단어' : list(dic.keys()),
    '빈도수' : list(dic.values())
})
df
```

```python
df.sort_values(by='빈도수',ascending=False)
```
