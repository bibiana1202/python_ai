```python
import numpy as np
import pandas as pd

perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])
```

```python
df = pd.read_csv("https://bit.ly/perch_csv")
df
```

```python
# 2차원 넘파이 배열로
df = df.to_numpy()
df
```

```python
from sklearn.model_selection import train_test_split

# 75% 훈련 25% 테스트
train_input, test_input, train_target, test_target = train_test_split(df,perch_weight)
```

```python
# 3개의 요인 : 길이 높이 가로 -> 강제로 늘리기 <특성공학>

from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures()
poly.fit([[2,3]]) # 2와3 가지고 만들수 있는 요인을 학습한다. 4,9,6,1(기본값),2,3
poly.transform([[2,3]])
```

```python
poly = PolynomialFeatures()
poly.fit(train_input)
train_poly = poly.transform(train_input)
train_poly.shape
```

```python
poly.get_feature_names()
```

```python
test_poly = poly.transform(test_input)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(test_poly,test_target)
```

```python
lr.score(test_poly,test_target)
```

```python
# 
poly = PolynomialFeatures(degree=5) # 최대 5제곱 까지 허용 한다.
poly.fit(train_input)
train_poly = poly.transform(train_input)
train_poly.shape
```

```python
# 3개엿던 요인이 56개로 뻥튀기
```

```python
test_poly = poly.transform(test_input)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(train_poly,train_target)
```

```python
lr.score(train_poly,train_target)
```

```python

lr.score(test_poly,test_target)
```

```python
# 과대 적합된 모습 1.0 / -64
```

```python
# 규제로 완화

# 표준화
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

ss.fit(train_poly) # 평균, 표준편차
train_scaled = ss.transform(train_poly) # (값 - 평균값) / 표준편차
test_scaled = ss.transform(test_poly)
```

```python
# 규제 모델 : 완벽하게 학습하는 것을 방해
# 릿지 모델 : 계수를 제곱한 값을 기준으로 
# 랏소 모델 : 계수의 절대값을 기준으로
# 선형모델에서 규제모델 추가됨

from sklearn.linear_model import Ridge

ridge = Ridge()
ridge.fit(train_scaled, train_target)
```

```python
ridge.score(test_scaled,test_target)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso.fit(train_scaled, train_target)
```

```python
lasso.score(test_scaled,test_target)
```
