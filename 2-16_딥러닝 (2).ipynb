```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2,4,6,8]) # 공부시간
y = np.array([65,80,75,95])

# 컴퓨터야 학습해~


# 기울기 절편 둘다 몰라~
a =0
b =0
lr =0.005 # 학습률

for i in range(10000):

  y_hat = a * x +b
  error = y_hat - y

  a_diff = sum(2 * error * x)
  b_diff = sum(2 * error * 1)

  a -= a_diff * lr
  b -= b_diff * lr

  print(f'{i+1}회 학습 // 기울기 {a} // 절편 {b}')




plt.scatter(x,y)
plt.plot(x, a*x+b , 'r')
plt.show()
```

```python
# 생각 보다 잘 안맞음 -> 요인이 2개 뿐이여서 강제로 늘려줘야

import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([2,4,6,8]) # 공부시간
x2 = np.array([1,2,1,3]) # 문제집 푼 갯수
y = np.array([65,80,75,95])

# 컴퓨터야 학습해~


# 기울기 절편 둘다 몰라~
a1 =0
a2 =0
b =0
lr =0.005 # 학습률

for i in range(10000):

  y_hat = a1 * x1 + a2 *x2 +b
  error = y_hat - y

  a1_diff = sum(2 * error * x1)
  a2_diff = sum(2 * error * x2)
  b_diff = sum(2 * error * 1)

  a1 -= a1_diff * lr
  a2 -= a2_diff * lr
  b -= b_diff * lr

  print(f'{i+1}회 학습 // 기울기1 {a1}// 기울기2 {a2} // 절편 {b}')

plt.scatter(x,y)
plt.plot(x, a1*x1 + a2*x2 + b , 'r')
plt.show()
```
