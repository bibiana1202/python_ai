### 문제3-2
로제떡볶이 1단계 맵기 3인분, 짜장떡볶이 2단계 맵기 2인분, 일반떡볶이 5단계 맵기 1인분을 주문하려고 한다. 

주문할 내역정보가 담긴 order 딕셔너리를 이용하여 내야 할 돈을 계산하려고 할 때, 

총 가격은 'price'라는 변수에 누적시켜 계산해야 할 총 가격을 출력하세요. 

(최대한 딕셔너리를 이용하기)

```python
menu = {'로제떡볶이' : {'맵기' : [1, 2, 3], '가격' : 7000},
        '투움바떡볶이' : {'맵기' : [1, 2, 3], '가격' : 7000},
        '일반떡볶이' : {'맵기' : [1, 2, 3, 4, 5], '가격' : 5000}}

order = {'로제떡볶이' : {'맵기' : 1, '갯수' : 3},
        '짜장떡볶이' : {'맵기' : 2, '갯수' : 2},
        '일반떡볶이' : {'맵기' : 5, '갯수' : 1}}

price = 0

m_k=menu.keys()
o_k=order.keys()

p = 0
for m in m_k:
  print(m)
  for o in o_k:
    if(m == o): 
      print('일치')
      o_kk=order[o].keys()
      for ok in o_kk:
        if(ok=='맵기'):
          print('맵기',order[o][ok])
          if(order[o][ok]>=3):
            print('4,5단계')
            p=menu[m]['가격']+500
            print('4,5가격',p)
          else:
            p=menu[m]['가격']  
            print('1,2,3가격',p)
        else:
          print('갯수',order[o][ok])
          p= p * order[o][ok]
          print(p)
      print('총가격',p)   
      price+=p 

print('최종가격',price)
```
