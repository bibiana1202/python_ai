info() 함수는 가변인자를 key값을 가진 딕셔너리 형태의 변수로 만들어주는 함수입니다.

```python
def info(**data):
    keys = data.keys()
    dic={}
    for k in keys:
      if(k=='name'):
        dic['이름']=data[k]
      elif(k=='age'):
        dic['나이']=data[k]
      elif(k=='number'):
        dic['번호']=data[k]
      else: print('오류')
    return dic 
```

```python
name = '홍길동'
age = 20
number = '010-1234-5678'

#한 줄로 코드 입력
info(**{'name' : name,'age' : age,'number' : number})
```

### 문제1
upper() 함수와 lower() 함수는 문자열을 각각 대문자, 소문자로 바꿔주는 내장된 함수입니다.

대문자를 소문자로 바꿔주는 함수를 직접 만들어보세요.

- 함수 이름 : my_lower
- 함수 내에서 eng 딕셔너리 사용
- lower() 외 함수 사용 가능

```python
def my_lower(str1) :
    eng = {'A': 'a', 'B': 'b', 'C': 'c', 'D': 'd', 'E': 'e', 'F': 'f', 'G': 'g', 'H': 'h', 'I': 'i', 'J': 'j', 'K': 'k',
            'L': 'l', 'M': 'm', 'N': 'n', 'O': 'o', 'P': 'p', 'Q': 'q', 'R': 'r', 'S': 's', 'T': 't', 'U': 'u', 'V': 'v',
            'W': 'w', 'X': 'x', 'Y': 'y', 'Z': 'z'}

    lower_str=''
    
    for str_i,str_v in enumerate(str1): # str
      for eng_i,eng_k in enumerate(eng): # eng
        if(str_v==eng_k): # 대문자
          lower_str+=eng[eng_k]
        elif(str_v==eng[eng_k] or str_v == ' '): # 소문자 공백
          lower_str+=str_v
          break
  
    return lower_str
    
```

```python
str3= my_lower('WelCome TO MY HOMe')
print(str3)
```

### 문제2-1
정수를 인자로 입력받으면 절대값으로 반환해주는 함수 to_abs()를 만들어보세요.

```python
def to_abs(num):
  if(num>0):
    return num
  elif(num == 0):
    return 0
  else:
    num = num * -1
    return num
```

### 문제2-2
to_abs()를 적절히 이용하여 입력받은 리스트의 원소를 절대값 크기로 내림차순 정렬해주는 함수를 만들어보세요.

```python
def sort_data(x_list):
  # 절대값 가진 딕셔너리 abs_dic
  abs_dic={}
  for i,v in enumerate(x_list):
    abs_dic[to_abs(v)]=v
  # 딕셔너리의 key 값을 내림차순 저장된 리스트 lower_list
  keys_list=list(abs_dic.keys())
  lower_list=[]

  for i_list,v_list in enumerate(keys_list):
    lower_list.append(v_list)

  lower_list.sort(reverse=True)

  # 절대값 크기로 내림차순 된 리스트 result_list
  result_list = []
    
  for i_list,v_list in enumerate(lower_list): 
    for i_dic,k_dic in enumerate(abs_dic):
      if(k_dic == v_list):
        result_list.append(abs_dic[k_dic])

  return result_list

```

```python
#함수를 생성한 뒤 실행하세요
x = [4, -5, 10, -15]
print(sort_data(x))
```

### 문제3
가변 매개변수를 이용해 입력받은 수의 평균을 구하는 average() 함수와

합을 구하는 sum() 함수를 구현하세요.

<실행>

print(my_average(1,2,3))

print(my_sum(1,2,3))

<출력 결과>

2

6

```python
#my_average

def my_average(*arg_list):
  sum_ma=0
  cnt_ma=0
  for i_list,v_list in enumerate(arg_list):
    sum_ma += v_list
    cnt_ma +=1
  
  return int(sum_ma/cnt_ma)



#my_sum
def my_sum(*arg):
  sum_ms =0
  for i,v in enumerate(arg):
    sum_ms += v
  
  return sum_ms



```

```python
print(my_average(*[1, 2, 3, 4, 5]))
print(my_sum(1, 2, 3, 4, 5))

print(my_average(1,2,3))

print(my_sum(1,2,3))
```

### 문제4
다음과 같이 정수가 중복으로 이루어져 있는 리스트가 여러 개 있을 때,

특정 리스트에 원하는 값을 인자로 입력하면 인덱스(위치)를 리턴해주는 함수를 만들어보세요.

<실행>

find_index(list2, 3)

<결과>

[2, 4, 7]

```python
#find_index

def find_index(list,find_v):
  result_list=[]
  for i_list,v_list in enumerate(list):
    if(find_v == v_list):
      result_list.append(i_list)
  
  return result_list

```
